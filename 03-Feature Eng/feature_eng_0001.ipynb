```python
#1a ideia
#normalizando variavel que se encontra fora de escala
df['wt_t'] = df['wt'] * 6
k = KMeans(n_clusters=4).fit(df[['wt_t', 'mpg']])
df['cluster'] = k.labels_
df.plot.scatter(x='wt', y='mpg', c='cluster', colormap='viridis')
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df['wt_01'] = scaler.fit_transform(df[['wt']])
df['mpg_01'] = scaler.fit_transform(df[['mpg']])

k = KMeans(n_clusters=4).fit(df[['wt_01', 'mpg_01']])
df['cluster'] = k.labels_
df.plot.scatter(x='wt', y='mpg', c='cluster', colormap='viridis')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Conjunto de dados de exemplo
dados = [50, 55, 60, 70, 90, 200]  # Note o outlier '200'

```

```python
# Calculando as medidas
media = np.mean(dados)
mediana = np.median(dados)
moda = stats.mode(dados)

# Exibindo os resultados
print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda[0]}")

# Visualizando os dados
plt.boxplot(dados)
plt.title('Boxplot dos Dados')
plt.show()
```

```python

import seaborn as sns

# Gerando um conjunto de dados com outliers
dados = np.random.normal(50, 15, 200)  # Média 50, desvio padrão 15, 200 pontos
print(dados)
#dados_com_outliers = np.append(dados, [150, 200, 300])  # Adicionando alguns outliers

# Criando o boxplot
plt.figure(figsize=(8,6))
sns.boxplot(dados)
plt.title('Boxplot com Outliers')
plt.show()
```

```python
# Criando dois conjuntos de dados
grupo1 = np.random.normal(60, 10, 100)
grupo2 = np.random.normal(50, 15, 100)

# Criando um DataFrame para facilitar a visualização com Seaborn
import pandas as pd
df = pd.DataFrame({'Grupo 1': grupo1, 'Grupo 2': grupo2})

# Visualizando o boxplot comparativo
plt.figure(figsize=(8,6))
sns.boxplot(data=df)
plt.title('Comparação de Grupos - Boxplot')
plt.show()
```

```python
# Gráfico de Dispersão (Scatter Plot)
plt.subplot(2, 1, 2)
# Adicionando um pouco de "jitter" (ruído) para os pontos não ficarem sobrepostos
x_values = np.random.normal(1, 0.04, size=len(grupo1))  # jitter para Grupo 1
plt.scatter(x_values, grupo1, alpha=0.6, color='blue', label='Grupo 1')

x_values = np.random.normal(2, 0.04, size=len(grupo2))  # jitter para Grupo 2
plt.scatter(x_values, grupo2, alpha=0.6, color='green', label='Grupo 2')

plt.xticks([1, 2], ['Grupo 1', 'Grupo 2'])
plt.title('Comparação de Grupos - Gráfico de Dispersão')
plt.ylabel('Valores')
plt.legend()

plt.tight_layout()
plt.show()
```
