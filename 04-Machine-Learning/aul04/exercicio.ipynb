```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/data/main/RiscoCredito%20-%20okk.csv')
test = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/data/main/RiscoCredito%20-%20prever2.csv')

train.head(4), test.head(4)
```

```python
from sklearn.tree import DecisionTreeClassifier

X_train = train[['Genero', 'GrauEscolaridade', 'Profissao', 'Renda']]
y_train = train['Inadimplente']



#treinar a arvode de decisão

modelo = DecisionTreeClassifier(max_depth=3)
modelo.fit(X_train, y_train)
```

```python
X_train, y_train
```

```python
#visualizar a árvore
from sklearn.tree import plot_tree
plt.figure(figsize=(8,4))
plot_tree(modelo, filled=True, feature_names=X_train.columns, class_names=['adimplente', 'inadimplente'])
plt.show()
```

```python
#medir assertividade

modelo.score(X_train, y_train)
```
