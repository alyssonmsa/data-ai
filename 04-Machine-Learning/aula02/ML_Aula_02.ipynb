```python
# import

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
treino = pd.read_csv('https://raw.githubusercontent.com/lcbjuk/ML/master/dados/notas%20-%20ok.csv')
```

```python
treino.head()
```

```python
reg_lin = LinearRegression()
reg_lin.fit(treino[['AP1', 'AP2']], treino['AP3'])
```

```python
print(reg_lin.coef_, reg_lin.intercept_, sep='\n')
```

```python
def prever_AP3(AP1, AP2):
  return reg_lin.coef_[0]*AP1 + reg_lin.coef_[1]*AP2 + reg_lin.intercept_
```

```python
print(prever_AP3(8, 8))
print(prever_AP3(2, 4))
print(prever_AP3(3.5, 8))
```

```python
teste = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/quant/master/notas-rl%20-%20test.csv')
```

```python
prev = reg_lin.predict(teste[['AP1', 'AP2']])
```

```python
metrics.r2_score(teste['AP3'], prev)
```

1. Será que existe alguma relação entre a cotação do dólar e a índice da Bovespa (IBOV)?
2. Essa relação é diretamente proporcional (positiva) ou negativa?
3. Qual o racional econômico que explica essa relação?
4. Crie então sua máquina para prever o futuro do dólar base na cotação do índice IBOV
5. A Bolsa hoje está em quanto? E o dólar? Se o IBOV chegar em 100 mil pontos, qual o valor do dólar previsto?

```python
df_usdbrl = pd.read_csv("https://raw.githubusercontent.com/lcbjrrr/data/main/usdbrl%20-%202024.csv")
df_usdbrl.head()
```

### 1. Será que existe alguma relação entre a cotação do dólar e a índice da Bovespa (IBOV)?

```python
df_usdbrl[["usdbrl", "bov"]].corr()
```

*R*: Sim, existe uma correlação negativa entre o valor do dolar e o indice da bovespa.

### 2.Essa relação é diretamente proporcional (positiva) ou negativa?
É uma relação negativa. De forma que quanto mais sobe o indice, mais o dolar desce.

### 3. Qual o racional econômico que explica essa relação?
Quanto mais alto o indice do bovespa, mais pessoas estão investindo na bolsa, de forma que, provavelmente, mais estrangeiros também estejam investindo na bovespa, trazendo mais dolar para o país, e por consequência, diminuindo a relação do valor do dolar em comparação com o real.

### 4. Crie então sua máquina para prever o futuro do dólar base na cotação do índice IBOV

```python
df_usdbrl.info()
```

```python
treino_usdbrl, teste_usdbrl = train_test_split(df_usdbrl, test_size=0.2)
reg_lin_bov = LinearRegression()
reg_lin_bov.fit(treino_usdbrl[["bov"]], treino_usdbrl["usdbrl"])
print(reg_lin_bov.coef_, reg_lin_bov.intercept_, sep='\n')
```

```python
reg_lin_bov_pred = reg_lin_bov.predict(teste_usdbrl[["bov"]])
metrics.r2_score(teste_usdbrl["usdbrl"], reg_lin_bov_pred)
```

### 5. A Bolsa hoje está em quanto? E o dólar? Se o IBOV chegar em 100 mil pontos, qual o valor do dólar previsto?

```python
# Criando um df com "bov" em 100 mil
df_bov_100mil = pd.DataFrame({"bov": [100]})
reg_lin_bov.predict(df_bov_100mil)
```
