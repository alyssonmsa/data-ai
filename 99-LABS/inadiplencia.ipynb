```python
#abrindo base inadimplencia na pasta bases
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
base = pd.read_excel('bases/BaseInadimplencia.xlsx')
base.head(20)
```

```python
fig, ax = plt.subplots()

ax.scatter(base.SaldoConta, base.SaldoInvestimento, c=base.Situacao)
x=[6,-6]
y=[-6,6]
ax.plot(x,y)
plt.show()
```

```python
# Preparando os dados
X = base[['SaldoConta', 'SaldoInvestimento']].values
y = base['Situacao'].values

# Separando os dados em duas classes (0 e 1) para plotar
class_0 = X[y == 0]
class_1 = X[y == 1]

# Grfico de disperso
plt.scatter(class_0[:, 0], class_0[:, 1], color='red', label='Classe 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='blue', label='Classe 1')

# Traar uma reta simples para separar os dados
x_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_vals = 0.5 * x_vals  # Esta  uma reta arbitrria (y = 0.5 * x)
plt.plot(x_vals, y_vals, color='green', label='Reta de separao')

plt.xlabel('SaldoConta')
plt.ylabel('SaldoInvestimento')
plt.legend()
plt.title('Grfico de Disperso com Reta de Separao')
plt.show()
```
