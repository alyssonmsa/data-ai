```python
import pandas as pd

# Criando um DataFrame com idades
dados = {'Idades': [15, 20, 22, 35, 42, 57, 63, 70, 81, 90]}
df = pd.DataFrame(dados)
df
```

```python
# Definindo os limites das categorias
categorias = pd.cut(df['Idades'], bins=[0, 18, 60, 100], labels=['Jovem', 'Adulto', 'Idoso'])
categorias
```

```python
# Adicionando a nova coluna categorizada ao DataFrame
df['Categoria'] = categorias

print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados de exemplo
tips = sns.load_dataset("tips")

# Criando o boxplot com Seaborn
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Boxplot com Seaborn")
plt.show()
```
