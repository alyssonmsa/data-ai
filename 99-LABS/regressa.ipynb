```python
import numpy as np
import matplotlib.pyplot as plt
from random import uniform

# 1. Gerar os valores de x e y (números de 0 a 4)
x = np.array([i for i in range(5)])
y = np.array([i for i in range(5)])

print(x, y)  # Exibe os arrays x e y no console para verificar os valores

# 2. Gerar pesos aleatórios (ws[0] é w0 e ws[1] é w1)
ws = np.array([round(uniform(-1, 1), 3) for i in range(2)])
print("Pesos ws:", ws)  # Exibe os pesos gerados aleatoriamente

# 3. Definir a função de hipótese (hipótese linear)
def hipotese(w0, w1, x):
    return w0 + w1 * x

# 4. Definir a função de erro quadrático médio (MSE - Mean Squared Error)
def MSE(w0, w1, x, y):
    return np.sum((hipotese(w0, w1, x) - y) ** 2) / len(x)

# 5. Plotar os pontos de dados reais (x, y)
plt.scatter(x, y, color='blue', label='Dados reais')

# 6. Plotar a linha da hipótese (reta ajustada pelos pesos ws)
plt.plot(x, hipotese(ws[0], ws[1], x), color='red', label='Hipótese')

# 7. Adicionar título e legendas no gráfico
plt.title("Hipótese Linear vs Dados Reais")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# 8. Mostrar o gráfico
plt.show()

# 9. Calcular e exibir o erro quadrático médio (MSE) com os pesos atuais
mse_value = MSE(ws[0], ws[1], x, y)
print("Erro Quadrático Médio (MSE):", mse_value)
```
